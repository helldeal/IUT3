- en apprentissage automatique, souvent on efface les lignes où il manque des données, faîtes le  => fonction *dropna(...)*




```python
import pandas as pd

# Charger le fichier CSV dans un dataframe
df = pd.read_csv('data/titanic.csv',sep="\t")

# Afficher les dimensions (shape) du dataframe
print("Dimensions du dataframe:", df.shape)

# Afficher les noms des colonnes du dataframe
print("\nNoms des colonnes du dataframe:", df.columns)

# Afficher la colonne nommée 'Age' du dataframe
print("\nColonne 'Age' du dataframe:")
print(df["Age"])

# Afficher les 5 premières lignes du dataframe de deux manières différentes
print("\n5 premières lignes (méthode 1):")
print(df.head())

print("\n5 premières lignes (méthode 2):")
print(df[:5])

# Afficher les 2 premières colonnes du dataframe
print("\n2 premières colonnes du dataframe:")
print(df.iloc[:, :2])

# Afficher le dataframe des lignes 25 à 50 et les colonnes à partir de la 3ème
print("\nLignes 25 à 50, colonnes à partir de la 3ème:")
print(df.iloc[24:49, 2:])

# Éliminer certaines colonnes (ne laisser que 'Pclass', 'Survived', 'Sex' et 'Age')
df = df[['Pclass', 'Survived', 'Sex', 'Age']]

# Afficher quelques statistiques de base
print("\nStatistiques de base:")
print(df.describe())

# Supprimer les lignes où il manque des données
df = df.dropna()

# Afficher les premières lignes du nouveau dataframe après suppression des lignes avec des valeurs manquantes
print("\nNouveau dataframe après suppression des lignes avec des valeurs manquantes:")
print(df.head())

```

### 2) Traitements plus complexes

- Donner l'âge minimum des passagers, l'âge maximum des passagers, l'âge moyen des passagers

- Afficher les lignes relatives aux passagers qui ont 21 ans

- Afficher les lignes relatives aux passagers qui ont l'âge maximum

- Afficher les lignes relatives aux passagers de sexe féminin qui ont plus de 30 ans

- Afficher un tableau avec les âges cumulés (10 personnes de 21 ans ...) puis trier ce tableau cumulé dans l'ordre descendant

- Des méthodes de dessin de graphique sont directement accessibles via les objets *pandas*. Dessiner le graphe relatif au tableau cumulé précédent (1 instruction)

- Dessiner un diagramme baton relatif au tableau cumulé précédent (1 instruction)

- Calculer l'âge moyen des passagers pour chaque classe de voyage (1 instruction)

- En regroupant les passagers par classe de voyage et par sexe, donner les moyennes de répartition des autres caractéristiques (1 seule instruction)

- Regrouper les données par classe de voyage et par sexe et calculer le nombre total de survivants dans chaque classe

- Calculer le pourcentage de survie pour chaque classe de voyage et chaque sexe.

- Calculer le pourcentage de survie en regroupant les données en fonction de la classe de voyage, du sexe et ceci pour des passagers âgés de moins de 18 ans

- Utiliser la fonction *pd.crosstab(...)* pour créer un tableau croisé montrant le nombre de survivants pour chaque classe de voyage et chaque sexe




```python
# Âge minimum, maximum et moyen des passagers
print("\nÂge minimum des passagers:", df['Age'].min())
print("Âge maximum des passagers:", df['Age'].max())
print("Âge moyen des passagers:", df['Age'].mean())

# Lignes relatives aux passagers qui ont 21 ans
print("\nPassagers qui ont 21 ans:")
print(df[df['Age'] == 21])

# Lignes relatives aux passagers qui ont l'âge maximum
print("\nPassagers qui ont l'âge maximum:")
print(df[df['Age'] == df['Age'].max()])

# Lignes relatives aux passagers de sexe féminin qui ont plus de 30 ans
print("\nPassagers de sexe féminin qui ont plus de 30 ans:")
print(df[(df['Sex'] == 'female') & (df['Age'] > 30)])

# Tableau avec les âges cumulés, trié dans l'ordre descendant
age_cumsum = df['Age'].value_counts().sort_index(ascending=False).cumsum()
print("\nÂges cumulés triés dans l'ordre descendant:")
print(age_cumsum)

# Dessiner le graphe relatif au tableau cumulé précédent
age_cumsum.plot(kind='line', title='Âges cumulés')

# Dessiner un diagramme bâton relatif au tableau cumulé précédent
age_cumsum.plot(kind='bar', title='Âges cumulés (diagramme bâton)')

# Âge moyen des passagers pour chaque classe de voyage
print("\nÂge moyen des passagers pour chaque classe de voyage:")
print(df.groupby('Pclass')['Age'].mean())

# Moyennes de répartition des autres caractéristiques par classe de voyage et par sexe
print("\nMoyennes de répartition par classe de voyage et par sexe:")
print(df.groupby(['Pclass', 'Sex']).mean())

# Nombre total de survivants par classe de voyage et par sexe
print("\nNombre total de survivants par classe de voyage et par sexe:")
print(df.groupby(['Pclass', 'Sex'])['Survived'].sum())

# Pourcentage de survie par classe de voyage et par sexe
survival_percentage = df.groupby(['Pclass', 'Sex'])['Survived'].mean() * 100
print("\nPourcentage de survie par classe de voyage et par sexe:")
print(survival_percentage)

# Pourcentage de survie pour les passagers âgés de moins de 18 ans par classe de voyage et par sexe
survival_percentage_under_18 = df[df['Age'] < 18].groupby(['Pclass', 'Sex'])['Survived'].mean() * 100
print("\nPourcentage de survie pour les passagers âgés de moins de 18 ans par classe de voyage et par sexe:")
print(survival_percentage_under_18)

# Tableau croisé montrant le nombre de survivants pour chaque classe de voyage et chaque sexe
cross_tab = pd.crosstab(df['Pclass'], df['Sex'], values=df['Survived'], aggfunc='sum', margins=True, margins_name='Total')
print("\nTableau croisé montrant le nombre de survivants pour chaque classe de voyage et chaque sexe:")
print(cross_tab)


```

### 3) Ensemble d'apprentissage, ensemble de test

- Charger le jeu de données Titanic à partir du .csv fourni

- Prétraitement des données (nous allons utiliser uniquement certaines colonnes: 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived')

- Gérer les valeurs manquantes (remplacer les valeurs manquantes de 'Age' par la moyenne)

- Convertir la colonne 'Sex' en une colonne binaire (0 pour *male*, 1 pour *female*)

- Découper le jeu de données en deux parties => X et Y:  
-- la première partie X correspond aux caractéristiques (donc toutes les colonnes sauf 'Survived')  
-- la seconde partie Y ne comprend que les données qui correspondent à la classe (la cible) qui ici est constituée de la colonne 'Survived'

- Diviser les données en respectant la répartition suivante:  80% pour l'apprentissage et 20% pour le test (regarder la fonction: *train_test_split(...)*)



```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Charger le jeu de données Titanic à partir du .csv fourni
df = pd.read_csv('data/titanic.csv',sep="\t")

# Prétraitement des données
selected_columns = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']
df = df[selected_columns]

# Gérer les valeurs manquantes (remplacer les valeurs manquantes de 'Age' par la moyenne)
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Convertir la colonne 'Sex' en une colonne binaire (0 pour male, 1 pour female)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Diviser le jeu de données en X (caractéristiques) et Y (cible)
X = df.drop('Survived', axis=1)
Y = df['Survived']

# Diviser les données en ensembles d'apprentissage et de test (80% pour l'apprentissage, 20% pour le test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Afficher les dimensions des ensembles d'apprentissage et de test
print("Dimensions de l'ensemble d'apprentissage X_train:", X_train.shape)
print("Dimensions de l'ensemble de test X_test:", X_test.shape)
print("Dimensions de l'ensemble d'apprentissage Y_train:", Y_train.shape)
print("Dimensions de l'ensemble de test Y_test:", Y_test.shape)

```
