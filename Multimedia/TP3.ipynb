```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
plt.close('all')

im = plt.imread('./cameraman.tif')


print(im.dtype)
print(im.max())
im = im / im.max()
im = 255 * im
im.astype(np.uint8)

kernel1 = np.array([[1,1,1],[1,1,1],[1,1,1]]) / 9
im1 = cv.filter2D(src=im, ddepth=-1,kernel=kernel1)
kernel2 = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]]) / 9
im2 = cv.filter2D(src=im, ddepth=-1,kernel=kernel2)

ig, axs = plt.subplots(1,2)
axs[0].imshow(im, cmap='gray')
axs[1].imshow(im1, cmap='gray')


```
